```
labels = df['5']
del df['5']
df.head()
```

```
df = pd.get_dummies(df)

df.head()
```

```
labels = pd.get_dummies(labels)
labels = labels.values
```

```
data = np.array(df)
labels = np.array(labels)
```

```
data=data.reshape(-1,88,1)
print(data.shape)
```

```
data = data.astype('float32') 
print(data.dtype) 

```

```
def Model():
    
    tf.keras.backend.clear_session()
    tf.random.set_seed(51)
    np.random.seed(51)


    model= tf.keras.models.Sequential([


       tf.keras.layers.LSTM(16,input_shape=(data.shape[1],1), return_sequences=True),

        tf.keras.layers.LSTM(16),
                 tf.keras.layers.Flatten(),

         tf.keras.layers.Dense(64, activation='relu'),
          tf.keras.layers.Dense(32, activation='relu'),
         tf.keras.layers.Dense(classes, activation='softmax'),

    ])

    model.compile(loss='categorical_crossentropy',  optimizer='adam',metrics = 'accuracy')
    
    return model
```

```
checkpoint_filepath = 'best_model'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1,
    save_best_only=True)
```

```
model= Model()
history = model.fit(data,labels,epochs=3,verbose=1,callbacks=[model_checkpoint_callback],validation_split=0.2,batch_size=32,shuffle=True)
```
